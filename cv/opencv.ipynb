```python
import cv2  
import numpy as np  
import IPython.display as dp
from google.colab.patches import cv2_imshow

# video load
v_controll = cv2.VideoCapture("/content/drive/MyDrive/data/h3.mp4")
v_o_bool, video_pic = v_controll.read()

def draw_conjour(pic):
    # to grey
    grey = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
    # to binary value
    binary = cv2.adaptiveThreshold(grey, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                    cv2.THRESH_BINARY_INV, 31, 17)
    # denoising
    matrix_p = np.ones((3,3), np.uint8)
    binary = cv2.morphologyEx(binary, cv2.MORPH_OPEN, matrix_p) 
    # contour construct
    contours_p, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    contour_draw = cv2.drawContours(binary, contours_p, -1, (255, 255, 255), 4)
    # rectangle bound
    x_b, y_b, width_b, height_b = cv2.boundingRect(contour_draw)
    pic = cv2.rectangle(pic, (x_b, y_b), (x_b + width_b, y_b + height_b), (0, 0, 255), 2)
    return pic

while v_o_bool:
    cv2_imshow(draw_conjour(video_pic))
    v_o_bool, video_pic = v_controll.read()
    dp.clear_output(True)
    cv2.waitKey(17)
```
